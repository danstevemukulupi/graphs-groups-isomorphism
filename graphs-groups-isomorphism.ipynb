### Matrices
***

```python
A = nx.to_numpy_array(G).astype(np.uint8)
A
```

```python
def mprint(A):
    for r in A:
        print(str(r)[1:-1])
        print()
```

```python
#from itertools import permutations
perms = it.permutations(range(len(A)))
print(list(perms))
```

```python
perms = it.permutations(range(len(A)))
I = np.identity(len(A)).astype(np.uint8)
for p in perms:
    mprint(I[list(p)])
```
